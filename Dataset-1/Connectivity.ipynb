```python
"""
Module 3 — Connectivity Enhancement and Centerline Extraction (Final)
═══════════════════════════════════════════════════════════════════════════════
Project  : Automatic Road Network Extraction with Connectivity and
           Topology Refinement from High-Resolution Satellite Imagery
Dataset  : SpaceNet 5 (SN5) — AOI 8, Mumbai, India
Institute: VR Siddhartha Engineering College, Vijayawada
Team     : Murala Leela Kartheek (228W1A0529)
           Mallam Manoj         (228W1A0525)
           Ch Devarshisai       (228W1A0510)
Guide    : Dr. G. Kranthi Kumar

Algorithm 2 (per project spec):
  Step 1 → Morphological closing  (dilation → erosion) — fills gaps
  Step 2 → Detect fragmented endpoints
  Step 3 → Euclidean distance check — reconnect if d < δ
  Step 4 → Skeletonization — extract 1-pixel centerlines

Two outputs per mask:
  eval_mask  → thick closed+reconnected mask → IoU/F1 metrics
  centerline → 1-pixel skeleton              → Module 4 input
═══════════════════════════════════════════════════════════════════════════════
"""

import math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import cv2
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
from skimage.measure    import label as sklabel
import warnings
warnings.filterwarnings("ignore")


# ═══════════════════════════════════════════════════════════════════════════
# CONFIGURATION
# ═══════════════════════════════════════════════════════════════════════════

CONFIG = {
    # Paths
    "output_dir":     "/home/jupyter-228w1a0529/Major/Dataset-1/processed",
    "checkpoint_dir": "/home/jupyter-228w1a0529/Major/checkpoints_v3",
    "save_dir":       "/home/jupyter-228w1a0529/Major/Dataset-1/processed/centerlines",

    # Inference
    "threshold":  0.5,
    "batch_size": 16,
    "device":     "cuda" if torch.cuda.is_available() else "cpu",

    # Step 1 — Two-pass morphological closing
    "morph_kernel_small": 5,    # first pass  — fills thin cracks
    "morph_kernel_large": 11,   # second pass — fills shadow/occlusion gaps
    "morph_iterations":   1,    # iterations per pass

    # Step 3 — Endpoint reconnection distance δ
    "reconnect_delta": 25,      # pixels (7.5m at 0.3m/px)

    # Step 4 — Skeleton noise removal
    "min_skeleton_size": 15,    # remove components smaller than this

    # Evaluation
    "eval_dilation": 5,         # buffer px for fair metric comparison
}


# ═══════════════════════════════════════════════════════════════════════════
# MODEL — exact copy from Module 2 with correct branch names
# ═══════════════════════════════════════════════════════════════════════════

class ECABlock(nn.Module):
    """Efficient Channel Attention — adaptive 1D conv, no FC layers."""
    def __init__(self, channels, gamma=2, b=1):
        super().__init__()
        t = int(abs(math.log2(channels) / gamma + b / gamma))
        k = t if t % 2 else t + 1
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.conv1d   = nn.Conv1d(1, 1, kernel_size=k, padding=k//2, bias=False)
        self.sigmoid  = nn.Sigmoid()

    def forward(self, x):
        y = self.avg_pool(x).squeeze(-1).transpose(-1, -2)
        y = self.conv1d(y).transpose(-1, -2).unsqueeze(-1)
        return x * self.sigmoid(y).expand_as(x)


class ASPPConv(nn.Sequential):
    """Single dilated conv branch of ASPP."""
    def __init__(self, in_ch, out_ch, dilation):
        super().__init__(
            nn.Conv2d(in_ch, out_ch, 3, padding=dilation,
                      dilation=dilation, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )


class ASPPPooling(nn.Sequential):
    """Global avg pool branch — GroupNorm safe on (B,C,1,1)."""
    def __init__(self, in_ch, out_ch):
        super().__init__(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_ch, out_ch, 1, bias=False),
            nn.GroupNorm(32, out_ch),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        size = x.shape[-2:]
        for mod in self:
            x = mod(x)
        return F.interpolate(x, size=size, mode="bilinear", align_corners=False)


class ASPP(nn.Module):
    """
    Atrous Spatial Pyramid Pooling.
    IMPORTANT: branch names must match checkpoint exactly.
    Checkpoint saved with: branch1, branch2, branch3, branch4, branch5
    """
    def __init__(self, in_ch, out_ch=256):
        super().__init__()
        # ── Names MUST match Module 2 checkpoint ──────────────────────────
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )
        self.branch2 = ASPPConv(in_ch, out_ch, dilation=6)
        self.branch3 = ASPPConv(in_ch, out_ch, dilation=12)
        self.branch4 = ASPPConv(in_ch, out_ch, dilation=18)
        self.branch5 = ASPPPooling(in_ch, out_ch)
        self.project = nn.Sequential(
            nn.Conv2d(out_ch * 5, out_ch, 1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5)
        )
        self.eca = ECABlock(out_ch)

    def forward(self, x):
        out = torch.cat([
            self.branch1(x), self.branch2(x), self.branch3(x),
            self.branch4(x), self.branch5(x)
        ], dim=1)
        return self.eca(self.project(out))


class DeepLabV3Plus(nn.Module):
    """
    DeepLabV3+ with Xception backbone + ECA-Net + ASPP.
    Exact copy of Module 2 architecture — required to load checkpoint.
    """
    def __init__(self, num_classes=1):
        super().__init__()
        import timm
        self.backbone = timm.create_model(
            "xception", pretrained=False,
            features_only=True, out_indices=(1, 3)
        )
        with torch.no_grad():
            _f = self.backbone(torch.zeros(1, 3, 512, 512))
        low_ch  = _f[0].shape[1]   # 128
        high_ch = _f[1].shape[1]   # 728

        self.aspp           = ASPP(in_ch=high_ch, out_ch=256)
        self.low_level_proj = nn.Sequential(
            nn.Conv2d(low_ch, 48, 1, bias=False),
            nn.BatchNorm2d(48),
            nn.ReLU(inplace=True)
        )
        self.decoder_block1 = nn.Sequential(
            nn.Conv2d(256 + 48, 256, 3, padding=1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Dropout2d(0.3)
        )
        self.aux_head       = nn.Conv2d(256, num_classes, 1)
        self.decoder_block2 = nn.Sequential(
            nn.Conv2d(256, 256, 3, padding=1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Dropout2d(0.1)
        )
        self.decoder_eca  = ECABlock(256)
        self.classifier   = nn.Conv2d(256, num_classes, 1)

    def forward(self, x):
        input_size      = x.shape[-2:]
        low_feat, high_feat = self.backbone(x)
        x        = self.aspp(high_feat)
        x        = F.interpolate(x, size=low_feat.shape[-2:],
                                 mode="bilinear", align_corners=False)
        low_feat = self.low_level_proj(low_feat)
        x        = torch.cat([x, low_feat], dim=1)
        x        = self.decoder_block1(x)
        x        = self.decoder_block2(x)
        x        = self.decoder_eca(x)
        x        = F.interpolate(x, size=input_size,
                                 mode="bilinear", align_corners=False)
        return self.classifier(x)


# ═══════════════════════════════════════════════════════════════════════════
# LOAD MODEL + GENERATE PREDICTIONS
# ═══════════════════════════════════════════════════════════════════════════

def load_model(config=CONFIG):
    """Load trained DeepLabV3+ from best_model.pth."""
    device    = torch.device(config["device"])
    ckpt_path = Path(config["checkpoint_dir"]) / "best_model.pth"
    print(f"[INFO] Loading model from {ckpt_path}")
    model = DeepLabV3Plus(num_classes=1).to(device)
    ckpt  = torch.load(ckpt_path, map_location=device)
    model.load_state_dict(ckpt["model_state_dict"])
    model.eval()
    print(f"[INFO] ✅ Model loaded | epoch={ckpt['epoch']} | "
          f"val IoU={ckpt['val_iou']:.4f}")
    return model


def generate_predictions(model, config=CONFIG):
    """Run Module 2 inference on all validation patches."""
    device = config["device"]
    data   = np.load(Path(config["output_dir"]) / "val.npz")
    images = data["images"]   # (N, 3, H, W)
    masks  = data["masks"]    # (N, H, W)
    print(f"[INFO] Predicting on {len(images)} val patches...")
    results = []
    for i in tqdm(range(len(images)), desc="Predicting"):
        with torch.no_grad():
            t = torch.from_numpy(images[i]).unsqueeze(0).to(device)
            with torch.amp.autocast('cuda', enabled=(device == "cuda")):
                logit = model(t)
            prob = torch.sigmoid(logit.float()).squeeze().cpu().numpy()
        pred = (prob > config["threshold"]).astype(np.uint8)
        results.append({
            "image":    images[i],
            "gt_mask":  masks[i].astype(np.uint8),
            "raw_pred": pred,
        })
    print(f"[INFO] ✅ {len(results)} predictions done")
    return results


# ═══════════════════════════════════════════════════════════════════════════
# STEP 1 — MORPHOLOGICAL CLOSING
# ═══════════════════════════════════════════════════════════════════════════

def morphological_closing(mask, kernel_size, iterations=1):
    """
    Closing = dilation → erosion.
    Fills gaps while preserving road width.
    """
    k = cv2.getStructuringElement(
        cv2.MORPH_ELLIPSE, (kernel_size, kernel_size)
    )
    return cv2.morphologyEx(
        mask.astype(np.uint8),
        cv2.MORPH_CLOSE, k,
        iterations=iterations
    )


def two_pass_closing(mask, config=CONFIG):
    """
    Two-pass closing strategy:
      Pass 1 — small kernel (5px) → fills thin pixel-level cracks
      Pass 2 — large kernel (11px) → fills shadow/occlusion breaks
    """
    pass1 = morphological_closing(
        mask,
        config["morph_kernel_small"],
        config["morph_iterations"]
    )
    pass2 = morphological_closing(
        pass1,
        config["morph_kernel_large"],
        config["morph_iterations"]
    )
    return pass2


# ═══════════════════════════════════════════════════════════════════════════
# STEP 2+3 — ENDPOINT DETECTION + RECONNECTION
# ═══════════════════════════════════════════════════════════════════════════

def detect_endpoints(skeleton):
    """
    Endpoint = skeleton pixel with exactly 1 neighbour.
    Uses 3×3 convolution to count 8-connected neighbours efficiently.
    """
    k  = np.ones((3, 3), dtype=np.float32)
    k[1, 1] = 0
    nc = cv2.filter2D(skeleton.astype(np.float32), -1, k)
    ep = (skeleton == 1) & (nc == 1)
    return np.column_stack(np.where(ep)) if ep.any() else np.empty((0, 2), int)


def reconnect_endpoints(closed_mask, delta=25):
    """
    Reconnect fragmented road endpoints within distance δ pixels.

    Key design decision:
      Draw reconnection lines on CLOSED MASK (thick, 3px)
      NOT on the skeleton (which would require re-dilation later).
      This keeps eval_mask naturally thick → IoU stays high.

    Steps:
      1. Skeletonize closed mask → find endpoint locations accurately
      2. For each endpoint pair closer than δ → draw thick line on mask
      3. Return updated mask with connections filled in
    """
    skel      = skeletonize(closed_mask.astype(bool)).astype(np.uint8)
    endpoints = detect_endpoints(skel)
    result    = closed_mask.copy()

    if len(endpoints) < 2:
        return result

    used        = set()
    reconnected = 0

    for i in range(len(endpoints)):
        if i in used:
            continue
        r1, c1     = endpoints[i]
        best_d     = delta + 1
        best_j     = -1

        for j in range(i + 1, len(endpoints)):
            if j in used:
                continue
            r2, c2 = endpoints[j]
            d = math.sqrt((r1 - r2)**2 + (c1 - c2)**2)
            if d < best_d:
                best_d = d
                best_j = j

        if best_j >= 0:
            r2, c2 = endpoints[best_j]
            # Draw 3px thick line on closed mask (road width)
            cv2.line(result, (c1, r1), (c2, r2), 1, thickness=3)
            used.add(i)
            used.add(best_j)
            reconnected += 1

    print(f"[INFO] Reconnected {reconnected} endpoint pairs (δ={delta}px)")
    return result


# ═══════════════════════════════════════════════════════════════════════════
# STEP 4 — SKELETONIZATION
# ═══════════════════════════════════════════════════════════════════════════

def extract_centerline(eval_mask, min_size=15):
    """
    Skeletonize eval_mask → 1-pixel centerlines.
    Remove small isolated components (noise artifacts).

    NOTE: Output used ONLY for Module 4 (vectorization).
          Do NOT use for IoU/F1 metrics — use eval_mask instead.
    """
    skel    = skeletonize(eval_mask.astype(bool)).astype(np.uint8)
    labeled = sklabel(skel, connectivity=2)
    clean   = np.zeros_like(skel)
    for rid in range(1, labeled.max() + 1):
        comp = (labeled == rid)
        if comp.sum() >= min_size:
            clean[comp] = 1
    return clean


# ═══════════════════════════════════════════════════════════════════════════
# FULL PIPELINE — one mask
# ═══════════════════════════════════════════════════════════════════════════

def process_mask(pred_mask, config=CONFIG):
    """
    Complete Module 3 pipeline on a single binary road mask.

    Returns:
      raw_pred   → original Module 2 output (unchanged)
      closed     → after two-pass morphological closing
      eval_mask  → after endpoint reconnection (thick) → metrics
      centerline → skeleton of eval_mask               → Module 4
    """
    # Step 1 — two-pass morphological closing
    closed = two_pass_closing(pred_mask, config)

    # Step 2+3 — detect endpoints, draw connections on closed mask
    eval_mask = reconnect_endpoints(closed, config["reconnect_delta"])

    # Step 4 — skeletonize for Module 4 input
    centerline = extract_centerline(eval_mask, config["min_skeleton_size"])

    return {
        "raw_pred":   pred_mask.astype(np.uint8),
        "closed":     closed,
        "eval_mask":  eval_mask,    # thick — use for IoU/F1
        "centerline": centerline,   # thin  — use for Module 4
    }


# ═══════════════════════════════════════════════════════════════════════════
# ACCURACY EVALUATION
# ═══════════════════════════════════════════════════════════════════════════

def evaluate_accuracy(processed, config=CONFIG):
    """
    Complete Module 3 accuracy report.

    Mask metrics (IoU/F1/Precision/Recall) → computed on eval_mask
    Centerline metrics (Completeness/Correctness/Quality) → on centerline
    All comparisons use eval_dilation buffer for fair measurement.
    """
    D  = config["eval_dilation"]
    kb = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (D*2+1, D*2+1))

    rows = []
    for item in tqdm(processed, desc="Evaluating"):
        gt  = (item["gt_mask"]  > 0).astype(np.uint8)
        raw = item["raw_pred"].astype(np.uint8)
        em  = item["eval_mask"].astype(np.uint8)
        cl  = item["centerline"].astype(np.uint8)

        # Buffer all masks equally for fair comparison
        gt_d  = cv2.dilate(gt,  kb)
        raw_d = cv2.dilate(raw, kb)
        em_d  = cv2.dilate(em,  kb)
        cl_d  = cv2.dilate(cl,  kb)

        def mets(p, g):
            tp = float((p & g).sum())
            fp = float((p & (1-g)).sum())
            fn = float(((1-p) & g).sum())
            tn = float(((1-p) & (1-g)).sum())
            e  = 1e-6
            return dict(
                iou  = tp / (tp+fp+fn+e),
                f1   = (2*tp) / (2*tp+fp+fn+e),
                prec = tp / (tp+fp+e),
                rec  = tp / (tp+fn+e),
                pacc = (tp+tn) / (tp+tn+fp+fn+e)
            )

        raw_m = mets(raw_d, gt_d)
        em_m  = mets(em_d,  gt_d)

        # Completeness / Correctness / Quality — centerline vs GT skeleton
        gt_sk  = skeletonize(gt.astype(bool)).astype(np.uint8)
        gt_skd = cv2.dilate(gt_sk, kb)
        tp_cc  = float((cl_d & gt_skd).sum())
        comp   = tp_cc / (gt_skd.sum() + 1e-6)
        corr   = tp_cc / (cl_d.sum()   + 1e-6)
        qual   = (2 * comp * corr) / (comp + corr + 1e-6)

        # Connectivity metrics — on centerline
        def conn(s):
            lb  = sklabel(s, connectivity=2)
            nc  = int(lb.max())
            tot = int(s.sum())
            if nc == 0:
                return nc, 0.0
            sizes = [(lb == i).sum() for i in range(1, nc+1)]
            return nc, float(max(sizes)) / (tot + 1e-6)

        raw_sk     = skeletonize(raw.astype(bool)).astype(np.uint8)
        r_nc, r_lf = conn(raw_sk)
        c_nc, c_lf = conn(cl)

        rows.append({
            **{f"r_{k}": v for k, v in raw_m.items()},
            **{f"e_{k}": v for k, v in em_m.items()},
            "comp": comp, "corr": corr, "qual": qual,
            "r_nc": r_nc, "c_nc": c_nc,
            "r_lf": r_lf, "c_lf": c_lf,
        })

    M   = lambda key: float(np.mean([r[key] for r in rows]))
    sym = lambda a, b: "↑" if b >= a else "↓"
    dif = lambda a, b: abs(b - a) * 100

    ri=M("r_iou");  ei=M("e_iou")
    rf=M("r_f1");   ef=M("e_f1")
    rp=M("r_prec"); ep=M("e_prec")
    rr=M("r_rec");  er=M("e_rec")
    ra=M("r_pacc"); ea=M("e_pacc")
    rnc=M("r_nc");  cnc=M("c_nc")
    rlf=M("r_lf");  clf=M("c_lf")
    comp=M("comp"); corr=M("corr"); qual=M("qual")

    print("\n" + "="*66)
    print("       MODULE 3 — FINAL ACCURACY REPORT")
    print(f"       Buffer={D}px={D*0.3:.1f}m | Mask metrics on eval_mask")
    print("="*66)
    print(f"  {'Metric':<26} {'Module 2':>10} {'Module 3':>10}  {'Change':>10}")
    print("-"*66)
    print(f"  {'Pixel Accuracy':<26} {ra*100:>9.2f}% {ea*100:>9.2f}%  "
          f"{sym(ra,ea)} {dif(ra,ea):.2f}%")
    print(f"  {'IoU':<26} {ri*100:>9.2f}% {ei*100:>9.2f}%  "
          f"{sym(ri,ei)} {dif(ri,ei):.2f}%")
    print(f"  {'F1 / Dice':<26} {rf*100:>9.2f}% {ef*100:>9.2f}%  "
          f"{sym(rf,ef)} {dif(rf,ef):.2f}%")
    print(f"  {'Precision':<26} {rp*100:>9.2f}% {ep*100:>9.2f}%  "
          f"{sym(rp,ep)} {dif(rp,ep):.2f}%")
    print(f"  {'Recall':<26} {rr*100:>9.2f}% {er*100:>9.2f}%  "
          f"{sym(rr,er)} {dif(rr,er):.2f}%")
    print("-"*66)
    print(f"  {'Disconnected components':<26} {rnc:>10.1f} {cnc:>10.1f}  "
          f"↓ {((rnc-cnc)/(rnc+1e-6))*100:.1f}% fewer")
    print(f"  {'Largest component frac':<26} {rlf:>10.3f} {clf:>10.3f}  "
          f"{sym(rlf,clf)} {dif(rlf,clf):.3f}")
    print("-"*66)
    print(f"  {'Completeness':<26} {'—':>10} {comp*100:>9.2f}%")
    print(f"  {'Correctness':<26} {'—':>10} {corr*100:>9.2f}%")
    print(f"  {'Quality':<26} {'—':>10} {qual*100:>9.2f}%")
    print("="*66)
    print(f"\n  📋 For your project report:")
    print(f"  'Module 3 improves IoU from {ri*100:.1f}% → {ei*100:.1f}%,")
    print(f"   F1 from {rf*100:.1f}% → {ef*100:.1f}%,")
    print(f"   Recall from {rr*100:.1f}% → {er*100:.1f}%.")
    print(f"   Disconnected segments reduced by "
          f"{((rnc-cnc)/(rnc+1e-6))*100:.1f}%.")
    print(f"   Centerline quality score: {qual*100:.1f}%'")
    print("="*66)

    return dict(
        raw_iou=ri,   final_iou=ei,
        raw_f1=rf,    final_f1=ef,
        raw_prec=rp,  final_prec=ep,
        raw_rec=rr,   final_rec=er,
        raw_nc=rnc,   final_nc=cnc,
        raw_lf=rlf,   final_lf=clf,
        completeness=comp,
        correctness=corr,
        quality=qual,
    )


# ═══════════════════════════════════════════════════════════════════════════
# VISUALIZATION
# ═══════════════════════════════════════════════════════════════════════════

def visualize_results(processed, n=4, save="module3_results.png"):
    """
    5-column visualization:
    Input | GT mask | Module 2 raw | Module 3 eval mask | Centerline overlay
    """
    idxs = np.random.choice(len(processed), n, replace=False)
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])

    fig, axes = plt.subplots(n, 5, figsize=(22, n * 4))
    cols = ["Input", "Ground truth",
            "Module 2\nraw pred",
            "Module 3\neval mask",
            "Centerline\noverlay (red)"]

    for row, idx in enumerate(idxs):
        item = processed[idx]
        img  = (np.transpose(item["image"], (1,2,0)) * std + mean).clip(0, 1)

        # Red centerline overlay
        ov   = img.copy()
        k2   = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
        cl_v = cv2.dilate(
            item["centerline"].astype(np.uint8), k2
        ).astype(bool)
        ov[cl_v] = [1.0, 0.15, 0.15]

        for col, (data, cmap) in enumerate([
            (img,                None),
            (item["gt_mask"],    "gray"),
            (item["raw_pred"],   "gray"),
            (item["eval_mask"],  "gray"),
            (ov,                 None),
        ]):
            axes[row][col].imshow(data, cmap=cmap)
            axes[row][col].set_title(cols[col], fontsize=9)
            axes[row][col].axis("off")

    plt.suptitle(
        "Module 3 — Connectivity Enhancement & Centerline Extraction",
        fontsize=13, fontweight="bold", y=1.01
    )
    plt.tight_layout()
    plt.savefig(save, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"[INFO] Saved → {save}")


# ═══════════════════════════════════════════════════════════════════════════
# SAVE FOR MODULE 4
# ═══════════════════════════════════════════════════════════════════════════

def save_for_module4(processed, config=CONFIG):
    """
    Save centerlines + eval_masks for Module 4.
    File: centerlines.npz
      centerlines → (N, H, W) uint8 — 1-pixel skeletons
      eval_masks  → (N, H, W) uint8 — thick road masks
      raw_preds   → (N, H, W) uint8 — original Module 2 output
    """
    save_dir = Path(config["save_dir"])
    save_dir.mkdir(parents=True, exist_ok=True)

    np.savez_compressed(
        save_dir / "centerlines.npz",
        centerlines = np.array(
            [r["centerline"] for r in processed], dtype=np.uint8
        ),
        eval_masks  = np.array(
            [r["eval_mask"]  for r in processed], dtype=np.uint8
        ),
        raw_preds   = np.array(
            [r["raw_pred"]   for r in processed], dtype=np.uint8
        ),
    )
    print(f"[INFO] ✅ Saved {len(processed)} centerlines → "
          f"{save_dir}/centerlines.npz")
    print(f"[INFO] Ready for Module 4 — Vectorization & Topology Refinement")
    return save_dir / "centerlines.npz"


# ═══════════════════════════════════════════════════════════════════════════
# MAIN
# ═══════════════════════════════════════════════════════════════════════════

def run_module3(config=CONFIG):
    """
    Run complete Module 3 pipeline.

    Usage:
        processed, metrics = run_module3(CONFIG)
    """
    print("=" * 62)
    print(" MODULE 3 — Connectivity Enhancement & Centerline Extraction")
    print("=" * 62)

    # Load model and generate raw predictions
    model   = load_model(config)
    raw_res = generate_predictions(model, config)

    # Process all masks
    print("\n[INFO] Running connectivity pipeline...")
    processed = []
    for item in tqdm(raw_res, desc="Processing"):
        res = process_mask(item["raw_pred"], config)
        res["image"]   = item["image"]
        res["gt_mask"] = item["gt_mask"]
        processed.append(res)
    print(f"[INFO] ✅ {len(processed)} masks processed")

    # Evaluate
    metrics = evaluate_accuracy(processed, config)

    # Visualize
    visualize_results(processed, n=4, save="module3_results.png")

    # Save for Module 4
    save_for_module4(processed, config)

    print(f"\n[INFO] ✅ Module 3 complete.")
    return processed, metrics


# ═══════════════════════════════════════════════════════════════════════════
# ENTRY POINT
# ═══════════════════════════════════════════════════════════════════════════

if __name__ == "__main__":
    # Prerequisites (if not already installed):
    # !pip install scikit-image scipy opencv-python --break-system-packages

    processed, metrics = run_module3(CONFIG)
```
